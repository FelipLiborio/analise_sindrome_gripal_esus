## Estratégia para tratamento de nulos

**Campos OBRIGATÓRIOS - Remover linhas se nulo:**
- `sexo`, `estado`, `municipio` - Dados demográficos básicos (0 nulos ✅)
- `dataNotificacao` - Data essencial (8 nulos - remover)
- `totalTestesRealizados` - Informação crítica (0 nulos ✅)
- `origem`, `excluido`, `validado` - Campos do sistema (0 nulos ✅)

**Campos CLÍNICOS - Manter nulos (informação ausente é válida):**
- `sintomas` (4), `racaCor` (108), `profissionalSaude` (4), `profissionalSeguranca` (175)
- `evolucaoCaso` (23210), `classificacaoFinal` (23929) - Casos em andamento
- `dataEncerramento` (24982) - Só preenche quando fecha o caso

**Campos de TESTAGEM - Manter nulos (nem todos fazem teste):**
- `codigoEstrategiaCovid` (7653), `codigoLocalRealizacaoTestagem` (7651)
- `codigoEstadoTeste1` (16720), `codigoTipoTeste1` (16720)
- `codigoFabricanteTeste1` (38068), `codigoResultadoTeste1` (20697)
- `dataColetaTeste1` (18831)

**Campos de VACINAÇÃO - Manter nulos (nem todos vacinados):**
- `codigoRecebeuVacina` (28085), `codigoDosesVacina` (53570)
- `dataPrimeiraDose` (57974), `dataSegundaDose` (69270)
- `codigoLaboratorioPrimeiraDose` (57369), `codigoLaboratorioSegundaDose` (68508)
- `lotePrimeiraDose` (57371), `loteSegundaDose` (68508)

**Campos PROBLEMÁTICOS:**
- `dataInicioSintomas` (60249) - **CRÍTICO!** Muitos nulos, mas é campo importante
- `idade` (39347) - **CRÍTICO!** Dado demográfico essencial
- `codigoContemComunidadeTradicional` (5687) - Aceitar nulo ou preencher como False?

```python
# Tratamento de nulos (preparação para dashboard e banco de dados)

print(f"Total de registros ANTES: {len(df)}")

# 1. Remover linhas com campos obrigatórios nulos
df = df.dropna(subset=['dataNotificacao'])
print(f"Após remover dataNotificacao nula: {len(df)}")

# 2. Preencher booleanos nulos como False (para dashboard não quebrar)
df['codigoContemComunidadeTradicional'] = df['codigoContemComunidadeTradicional'].fillna(False).infer_objects(copy=False)
df['profissionalSaude'] = df['profissionalSaude'].fillna(False).infer_objects(copy=False)
df['profissionalSeguranca'] = df['profissionalSeguranca'].fillna(False).infer_objects(copy=False)
df['codigoRecebeuVacina'] = df['codigoRecebeuVacina'].fillna(False)

# 3. Categorias texto: "Não Informado" para visualização
df["racaCor"] = df["racaCor"].fillna("Ignorado")
df['evolucaoCaso'] = df['evolucaoCaso'].fillna('Em Andamento')
df['classificacaoFinal'] = df['classificacaoFinal'].fillna('Não Classificado')
df['sintomas'] = df['sintomas'].fillna('Assintomático')

# 4. Códigos categóricos: 0 = "Não Informado" (tabelas de domínio)
df['codigoEstrategiaCovid'] = df['codigoEstrategiaCovid'].fillna(0)
df['codigoLocalRealizacaoTestagem'] = df['codigoLocalRealizacaoTestagem'].fillna(0)
df['codigoDosesVacina'] = df['codigoDosesVacina'].fillna(0)
df['codigoFabricanteTeste1'] = df['codigoFabricanteTeste1'].fillna(0)
df['codigoResultadoTeste1'] = df['codigoResultadoTeste1'].fillna(0)
df['codigoEstadoTeste1'] = df['codigoEstadoTeste1'].fillna(0)
df['codigoTipoTeste1'] = df['codigoTipoTeste1'].fillna(0)
df['codigoLaboratorioPrimeiraDose'] = df['codigoLaboratorioPrimeiraDose'].fillna(0)
df['codigoLaboratorioSegundaDose'] = df['codigoLaboratorioSegundaDose'].fillna(0)

# 5. Lotes de vacina: "Não Informado"
df['lotePrimeiraDose'] = df['lotePrimeiraDose'].fillna('Não Informado')
df['loteSegundaDose'] = df['loteSegundaDose'].fillna('Não Informado')

# 6. Regra especial: dataInicioSintomas
# Se nula, usar dataNotificacao (casos assintomáticos)
df.loc[df['dataInicioSintomas'].isna(), 'dataInicioSintomas'] = df.loc[df['dataInicioSintomas'].isna(), 'dataNotificacao']

# 7. MANTER NULL: Datas opcionais (linha do tempo epidemiológica)
# - dataPrimeiraDose, dataSegundaDose, dataColetaTeste1, dataEncerramento
# - idade (dado demográfico ausente é válido)

print(f"\n✅ Tratamento epidemiológico concluído!")
print(f"Registros mantidos: {len(df)}")
print(f"\nNulos restantes (apenas datas/idade opcionais):")
print(df.isnull().sum()[df.isnull().sum() > 0])
```
